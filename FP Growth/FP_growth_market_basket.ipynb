```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/Market_Basket_Optimisation.csv',header=None)
```

```python
data.head()
```

```python
transactions=[]
```

```python
data.shape
```

```python
for i in range(0,7501):
  transactions.append([str(data.values[i,j]) for j in range(0,20)])
```

```python
transactions
```

```python
import pyfpgrowth as fpg
```

```python
patterns=fpg.find_frequent_patterns(transactions,2)
```

```python
rules=fpg.generate_association_rules(patterns,0.7)
```
